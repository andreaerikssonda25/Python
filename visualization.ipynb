```python
import requests
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("greek recipe calories.csv")

df.info()
```

```python
#Create a bar chart showing the calories per recipe?
#Convert "calories" column to numeric
df["calories"] = pd.to_numeric(df["calories"], errors="coerce")

plt.figure(figsize=(10, 6))
plt.bar(df["title"], df["calories"], color="blue")
plt.title("Greek Recipes")
plt.xticks(rotation=90)
plt.xlabel("Recipes")
plt.ylabel("Calories")
plt.show()
```

```python
#Filter recipes with over 400 calories
high_calorie_recipes = df[df["calories"] > 400]
        
#Order by title, from A to Z
high_calorie_recipes_sorted = high_calorie_recipes.sort_values(by="title")
        
print(high_calorie_recipes_sorted[["title", "calories"]])
```

```python
#What would be the mean amount of calories from the above data set? Round to 2 decimal places
mean_calories = round(df["calories"].mean(), 2)
print(mean_calories)
```

```python
#Plot the high calorie recipes (400+) in a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(high_calorie_recipes["title"], high_calorie_recipes["calories"], color="orange")
plt.title("High Cal Greek Recipes")
plt.xticks(rotation=90)
plt.xlabel("Recipes")
plt.ylabel("Calories")
plt.show()
```

```python
#Which recipe out of the entire bunch has the lowest amount of calories?
smallest_calorie_recipe = df[df["calories"] == df["calories"].min()].iloc[0]
print(smallest_calorie_recipe)
```
